```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import time
import random
```

```python
links = []
```

```python
def get_jobs(url):
    jobs_title = []
    company = []
    job_location = []
    job_image = []
    job_date = []
    job_active = []
    
    for i in range(0, 75, 25):
        html = urlopen(url+"&start="+str(i))
        soup = BeautifulSoup(html, 'lxml')

        job_results = soup.find("ul", attrs={"class":"jobs-search__results-list"}).find_all("li")
        
        for job in job_results:
            links.append(job.find("a", attrs={"class":"base-card__full-link"}))
        
        jobs_info = []
        for job in job_results:
            jobs_info.append(job.find("div", attrs={"class":"base-search-card__info"}))
            
        for info in jobs_info:
            jobs_title.append(info.find("h3", attrs={"class":"base-search-card__title"}).text.strip())
            
        for info in jobs_info:
            company.append(info.find("h4", attrs={"class":"base-search-card__subtitle"}).text.strip())
            
        job_meta = []
        for meta in jobs_info:
            job_meta.append(meta.find("div", attrs={"class":"base-search-card__metadata"}))
            
        for location in job_meta:
            job_location.append(location.find("span", attrs={"class":"job-search-card__location"}).text.strip())
            
        for date in job_meta:
            job_date.append(date.find("time").text.strip())
            
        job_images = soup.find_all("img", {"class": "artdeco-entity-image--square-4"})
        for img in job_images:
            job_image.append(img['data-delayed-url'])
            
        time.sleep(6+random.random()*10)
        
    return pd.DataFrame({"Company": company, "Job Title": jobs_title, "Job Location":job_location, "Job Date": job_date, "Job Image URL": job_image})
```

```python
jobs_data_ml = get_jobs("https://www.linkedin.com/jobs/search/?keywords=machine%20learning%20jobs")
jobs_data_ml
```

```python
urls = []
for link in links:
    try:
        urls.append(link["href"])
    except:
        urls.append("")
urls = [i for i in urls if i]
```

```python
job_title = []
job_company = []
job_location = []
job_image = []
job_date = []
job_describtion = []
job_tags = []
job_requirements = []
```

```python
def get_job_data(_all):
    
    for url in _all:
        html = urlopen(url)
        soup = BeautifulSoup(html, 'lxml')
        
        job_title.append(soup.find("h1").text.strip())
        cname = soup.find("a", attrs= {"class":"topcard__org-name-link"}).text.strip()
        job_company.append(cname)
        job_location.append(soup.find("span", attrs= {"class":"topcard__flavor topcard__flavor--bullet"}).text.strip())
        job_image.append(soup.find("img", attrs={"class": "sub-nav-cta__image"})["data-delayed-url"])
        job_date.append(soup.find("span", attrs= {"class":"posted-time-ago__text"}).text.strip())
        more = []
        m = soup.find_all("span", attrs={"class":"description__job-criteria-text description__job-criteria-text--criteria"})
        for e in m:
            more.append(e.text.strip())
        job_tags.append(more)
        requirements_ = []
        requires = soup.find("div", attrs={"class":"show-more-less-html__markup"}).find_all("li")
        for li in requires:
            requirements_.append(li.text.strip())
        job_requirements.append(requirements_)
        about = soup.find("div", attrs={"class":"show-more-less-html__markup"}).get_text(separator="\n ").split("\n")
        about = [i.strip() for i in about]
        about = [i for i in about if i]
        about = [i for i in about if not ( i.startswith("About") or i.startswith("#") )]
        job_describtion.append(about[0])
        time.sleep(7+random.random()*10)
    return pd.DataFrame({"Company": job_company, "Job Title": jobs_title, "Job Location":job_location, "Job Date": job_date, "Job Image URL": job_image, "Describtion": job_describtion, "Tags": job_tags, "Requirements": job_requirements})
```

```python
all_data = get_job_data(urls)
all_data
```
